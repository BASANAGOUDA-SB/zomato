```python

# Read Country-code dataset
zom_country = pd.read_excel('country-code.xlsx')
zom_country.head()
```

```python
zom.columns
```

```python
## merge both dataset
final_zom=pd.merge(zom,zom_country,on ='Country Code',how='left')
final_zom.head()
```

```python
final_zom.dtypes
```

```python
final_zom.columns
```

```python
## identifying all diff country names
country_names = final_zom.Country.value_counts().index
country_names
```

```python
## seperating all values 
country_val = final_zom.Country.value_counts().values
country_val
```

```python
## creating pie hart between top 3 countries and its valuecounts
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_zom.columns
```

```python
ratings=ratings = final_zom.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating count"})
```

```python
plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x = 'Aggregate rating', y = 'Rating text', data = ratings)
```

```python
#Basically we use it for plotting catogrical variables.
# On y axis we have frequency. white-1, red- 7, yellow-5 , orange-10 and so on..
sns.countplot(x = 'Rating color', data = ratings, palette =['blue', 'Red', 'Orange', 'Yellow', 'Green','Green'] )
```
